```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import yfinance as yf
import pickle

# Download data for the S&P 500 index from 2020 to 2022
ticker = "NDX" #"SPY"
start_date = "1973-01-01"
end_date = "2023-05-25"
sp500_data = yf.download(ticker, start=start_date, end=end_date)

# Print the first 5 rows of the data
print(sp500_data.head())
# Plot the adjusted close price
#sp500_data["Adj Close"].plot()
plt.plot(np.log(sp500_data["Adj Close"]))

# Extract the adjusted close prices
adj_close = sp500_data["Adj Close"]

# Calculate the logarithm of the adjusted close prices
log_adj_close = np.log(adj_close)

# Create an array of x-values corresponding to the indices of the data points
x = np.arange(len(log_adj_close))

# Fit a linear regression line to the data
coefficients = np.polyfit(x, log_adj_close, 1)
line_of_best_fit = np.polyval(coefficients, x)

# Plot the adjusted close prices and the line of best fit
plt.plot(log_adj_close, label="Adjusted Close")
plt.plot(sp500_data.index, line_of_best_fit, label="Line of Best Fit")
plt.xlabel("Index")
plt.ylabel("Log Adjusted Close")
plt.title("Line of Best Fit for Adjusted Close Prices")
plt.legend()

# Display the plot
plt.show()
```
